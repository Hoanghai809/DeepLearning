## BTVN 1:
## cho y = 5x^5 + 6x^3 - 3x + 1
## cho biet do doc cua da thuc tren o diem nao

```python
import torch
import numpy as np

def calculate_gradient_at_point(x_value):
    """
    Tính độ dốc của hàm y = 5x^5 + 6x^3 - 3x + 1 tại điểm x_value
    """
    x = torch.tensor(x_value, dtype=torch.float32, requires_grad=True)
    y = 5 * x**5 + 6 * x**3 - 3 * x + 1
    y.backward()
    return x.grad.item()

def find_points_with_specific_gradient(target_gradient):
    """
    Tìm tất cả các điểm x mà tại đó độ dốc = target_gradient
    """
    # Giải phương trình: 25x^4 + 18x^2 - 3 = target_gradient
    # Tức là: 25x^4 + 18x^2 - (3 + target_gradient) = 0
    
    # Đặt t = x^2, phương trình trở thành: 25t^2 + 18t - (3 + target_gradient) = 0
    a = 25
    b = 18
    c = -(3 + target_gradient)
    
    discriminant = b**2 - 4*a*c
    
    if discriminant < 0:
        return []  # Không có nghiệm thực
    
    solutions = []
    
    # Tính nghiệm t
    sqrt_disc = np.sqrt(discriminant)
    t1 = (-b + sqrt_disc) / (2*a)
    t2 = (-b - sqrt_disc) / (2*a)
    
    # Kiểm tra và tính nghiệm x từ t
    if t1 >= 0:
        x1 = np.sqrt(t1)
        x2 = -np.sqrt(t1)
        solutions.extend([x1, x2])
    
    if t2 >= 0 and abs(t2 - t1) > 1e-10:  # Tránh trùng lặp
        x3 = np.sqrt(t2)
        x4 = -np.sqrt(t2)
        solutions.extend([x3, x4])
    
    # Làm tròn và loại bỏ trùng lặp
    solutions = list(set([round(x, 6) for x in solutions]))
    return sorted(solutions)

def main():
    print("=" * 60)
    print("TÍNH ĐỘ DỐC CỦA HÀM: y = 5x^5 + 6x^3 - 3x + 1")
    print("=" * 60)
    
    # 1. Tính độ dốc tại một số điểm cụ thể
    print("\n1. ĐỘ DỐC TẠI CÁC ĐIỂM CỤ THỂ:")
    
    test_points = [-2, -1.5, -1, -0.5, 0, 0.5, 1, 1.5, 2]
    
    for x_val in test_points:
        gradient = calculate_gradient_at_point(x_val)
        print(f"   Tại x = {x_val:5.1f}: dy/dx = {gradient:8.4f}")
    
    # 2. Tìm điểm có độ dốc = 0 (điểm cực trị)
    print("\n2. ĐIỂM CỰC TRỊ (độ dốc = 0):")
    zero_gradient_points = find_points_with_specific_gradient(0)
    
    if zero_gradient_points:
        print(f"   Tìm thấy {len(zero_gradient_points)} điểm:")
        for i, x in enumerate(zero_gradient_points, 1):
            y = 5*x**5 + 6*x**3 - 3*x + 1
            gradient_check = calculate_gradient_at_point(x)
            print(f"   Điểm {i}: x = {x:7.4f}, y = {y:7.4f}, dy/dx = {gradient_check:.10f}")
    else:
        print("   Không tìm thấy điểm nào có độ dốc = 0")
    
    # 3. Tìm điểm với độ dốc cụ thể khác
    print("\n3. TÌM ĐIỂM VỚI ĐỘ DỐC KHÁC:")
    
    target_gradients = [-5, -3, -1, 10, 50, 100]
    
    for target in target_gradients:
        points = find_points_with_specific_gradient(target)
        
        if points:
            print(f"   Độ dốc = {target:4}: Tìm thấy {len(points)} điểm")
            for x in points:
                y = 5*x**5 + 6*x**3 - 3*x + 1
                print(f"        x = {x:7.4f}, y = {y:7.4f}")
        else:
            print(f"   Độ dốc = {target:4}: Không có điểm nào")
    
    # 4. Tính độ dốc nhỏ nhất có thể
    print("\n4. PHÂN TÍCH ĐỘ DỐC:")
    
    # Đạo hàm: 25x^4 + 18x^2 - 3
    # Đây là hàm bậc 4, hệ số dương, nên có giá trị nhỏ nhất tại x = 0
    min_gradient = calculate_gradient_at_point(0)
    print(f"   Độ dốc nhỏ nhất có thể: {min_gradient} (tại x = 0)")
    print(f"   Công thức tổng quát: dy/dx = 25x^4 + 18x^2 - 3")
    print(f"   Với mọi x, dy/dx ≥ {min_gradient}")
    
    # 5. Nhập điểm từ người dùng
    print("\n5. TÍNH ĐỘ DỐC TẠI ĐIỂM TÙY CHỌN:")
    
    while True:
        try:
            user_input = input("\nNhập giá trị x (hoặc 'q' để thoát): ")
            
            if user_input.lower() == 'q':
                break
                
            x_val = float(user_input)
            gradient = calculate_gradient_at_point(x_val)
            y_val = 5*x_val**5 + 6*x_val**3 - 3*x_val + 1
            
            print(f"   Tại x = {x_val}:")
            print(f"   - Giá trị hàm số: y = {y_val}")
            print(f"   - Độ dốc: dy/dx = {gradient}")
            print(f"   - Công thức kiểm tra: 25*({x_val})⁴ + 18*({x_val})² - 3 = {25*x_val**4 + 18*x_val**2 - 3}")
            
        except ValueError:
            print("Vui lòng nhập số hợp lệ!")
        except Exception as e:
            print(f"Lỗi: {e}")

if __name__ == "__main__":
    main()
```

## # BTVN 2:
## tao mot tensor ban dau co gia tri la 2
## dinh nghia ham so va tinh gradiant
## y = x^3 + 2x^2 + 5x + 1
## hay tinh dy/dx tai gia tri cua x 
## dung phuong phap Gradiant Descent voi learning rate alpha = 0.01 de cap nhat gia tri cua x sau 10 vong lap

```python
import torch

# BTVN 2: Gradient Descent với y = x^3 + 2x^2 + 5x + 1

# 1. Tạo tensor ban đầu
x = torch.tensor(2.0, requires_grad=True)

# 2. Định nghĩa hàm số
def f(x):
    return x**3 + 2*x**2 + 5*x + 1

# 3. Gradient Descent
learning_rate = 0.01
iterations = 10

print("Gradient Descent: y = x^3 + 2x^2 + 5x + 1")
print("=" * 50)
print(f"Bắt đầu với x = {x.item():.4f}")

for i in range(iterations):
    # Tính giá trị hàm số
    y = f(x)
    
    # Tính gradient
    y.backward()
    
    # In kết quả
    print(f"Vòng {i+1:2d}: x = {x.item():8.4f}, y = {y.item():8.4f}, dy/dx = {x.grad.item():8.4f}")
    
    # Cập nhật x (gradient descent)
    with torch.no_grad():
        x -= learning_rate * x.grad
    
    # Reset gradient
    x.grad.zero_()

print("=" * 50)
print(f"Kết thúc: x = {x.item():.4f}, y = {f(x).item():.4f}")
```

## # BTVN 3:
## tao mot tap du lieu gia lap voi x la so h hoc ngau nhien tu 1 - 10 va y la so dem duoc tinh theo cong thuc 
## y = 3x + 5 + noise voi noise la mot gia tri ngau nhien nho

## 1 khoi tao tham so W va b ngau nhien voi requies_grad = True
## 2 tinh MSE
## 3 tinh Gradiant
## 4 cap nhat W va b theo phuong phap Gradiant Descent voi learning rate alpha = 0.01
## 5 lap lai 100 lan de xem su hoi tu cua mo hinh the nao

```python
import torch
import numpy as np

# Tạo dữ liệu giả lập
np.random.seed(42)
x_data = np.random.uniform(1, 10, 50)          # Số giờ học: 50 điểm từ 1-10 giờ
noise = np.random.normal(0, 1.5, 50)           # Noise nhỏ
y_data = 3 * x_data + 5 + noise                # y = 3x + 5 + noise

# Chuyển sang tensor
x_tensor = torch.tensor(x_data, dtype=torch.float32)
y_tensor = torch.tensor(y_data, dtype=torch.float32)

# 1. Khởi tạo tham số W và b ngẫu nhiên
W = torch.randn(1, requires_grad=True)
b = torch.randn(1, requires_grad=True)

print("=" * 60)
print("BÀI TẬP 3: HỒI QUY TUYẾN TÍNH")
print("=" * 60)
print(f"Dữ liệu: 50 điểm, x ∈ [1, 10] giờ học")
print(f"Công thức thực: y = 3x + 5 + noise")
print(f"Khởi tạo: W = {W.item():.4f}, b = {b.item():.4f}")
print("=" * 60)

# 2. Hàm tính MSE
def mse_loss(y_pred, y_true):
    return torch.mean((y_pred - y_true) ** 2)

# Gradient Descent
learning_rate = 0.01
epochs = 100

print("\nTIẾN TRÌNH HUẤN LUYỆN:")
print("-" * 60)

for epoch in range(epochs):
    # Dự đoán
    y_pred = W * x_tensor + b
    
    # Tính MSE loss
    loss = mse_loss(y_pred, y_tensor)
    
    # 3. Tính gradient (tự động với backward)
    loss.backward()
    
    # 4. Cập nhật W và b với Gradient Descent
    with torch.no_grad():
        W -= learning_rate * W.grad
        b -= learning_rate * b.grad
        
        # Reset gradient
        W.grad.zero_()
        b.grad.zero_()
    
    # In kết quả mỗi 25 epoch
    if (epoch + 1) % 25 == 0:
        print(f"Epoch {epoch+1:3d}: W = {W.item():6.3f}, b = {b.item():6.3f}, Loss = {loss.item():7.3f}")

# Kết quả cuối cùng
print("-" * 60)
print("\nKẾT QUẢ SAU 100 EPOCH:")
print(f"Giá trị thực tế: W = 3.0, b = 5.0")
print(f"Giá trị học được: W = {W.item():.3f}, b = {b.item():.3f}")
print(f"Loss cuối cùng: {loss.item():.3f}")

# Tính R-squared để đánh giá
y_pred_final = W * x_tensor + b
ss_res = torch.sum((y_tensor - y_pred_final) ** 2)
ss_tot = torch.sum((y_tensor - torch.mean(y_tensor)) ** 2)
r_squared = 1 - (ss_res / ss_tot)

print(f"R² Score: {r_squared.item():.4f} (càng gần 1 càng tốt)")

# Dự đoán thử
print("\nVÍ DỤ DỰ ĐOÁN:")
test_hours = [2, 5, 8, 10]
for hours in test_hours:
    predicted = W.item() * hours + b.item()
    actual = 3 * hours + 5  # Không có noise
    print(f"  {hours} giờ học: Dự đoán = {predicted:.1f} điểm, Thực tế ≈ {actual} điểm")
```

## Resahpe voi view va view as

```python
import torch

# Tạo tensor ban đầu
original = torch.arange(12)  # tensor([0, 1, 2, ..., 11])
print("Tensor ban đầu:")
print(original)
print(f"Kích thước: {original.shape}\n")

# 5.1 Sử dụng view() để reshape
# Reshape thành 3x4
reshaped_3x4 = original.view(3, 4)
print("Reshape thành 3x4 (view):")
print(reshaped_3x4)
print(f"Kích thước: {reshaped_3x4.shape}\n")

# Reshape thành 2x2x3
reshaped_2x2x3 = original.view(2, 2, 3)
print("Reshape thành 2x2x3 (view):")
print(reshaped_2x2x3)
print(f"Kích thước: {reshaped_2x2x3.shape}\n")

# Sử dụng -1 để tự động tính toán kích thước
auto_reshape = original.view(2, -1)  # Tự động tính: 12/2 = 6 → 2x6
print("Reshape với -1 (2x-1):")
print(auto_reshape)
print(f"Kích thước: {auto_reshape.shape}\n")

# 5.2 Sử dụng view_as() để reshape theo tensor khác
target_shape = torch.zeros(4, 3)  # Tensor mẫu 4x3
reshaped_as = original.view_as(target_shape)
print("Reshape theo tensor mẫu 4x3 (view_as):")
print(reshaped_as)
print(f"Kích thước: {reshaped_as.shape}\n")

# 5.3 So sánh với reshape()
# reshape() là alias của view() (trong hầu hết trường hợp)
reshaped_method = original.reshape(3, 4)
print("Reshape với reshape() method (3x4):")
print(reshaped_method)
print(f"Kích thước: {reshaped_method.shape}\n")

# 5.4 Kiểm tra chia sẻ bộ nhớ
print("KIỂM TRA CHIA SẺ BỘ NHỚ:")
print(f"original.data_ptr() = {original.data_ptr()}")
print(f"reshaped_3x4.data_ptr() = {reshaped_3x4.data_ptr()}")
print(f"Chia sẻ bộ nhớ? {original.data_ptr() == reshaped_3x4.data_ptr()}\n")

# 5.5 Thay đổi giá trị
print("THAY ĐỔI GIÁ TRỊ:")
reshaped_3x4[0, 0] = 99
print("Sau khi đặt reshaped_3x4[0,0] = 99:")
print(f"Original tensor: {original[0]}")
print(f"Reshaped tensor[0,0]: {reshaped_3x4[0,0]}")
```
